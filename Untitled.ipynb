```python
inp = L.Input([None, None, 3])
temp = n = L.Conv2D(64, (3, 3), padding='same')(inp)

for i in range(4):
    nn = L.Deconv2D(64, (3, 3), activation='elu', padding='same')(n)
    nn = L.Add()([n, nn])
    n = nn

conv2 = L.Deconv2D(64, (3, 3), activation='elu', padding='same')(n)
mer = L.Add()([conv2, temp])

U_s2 = n = L.UpSampling2D((2, 2))(mer)


for i in range(4):
    nn = L.Deconv2D(64, (3, 3), activation='elu', padding='same')(n)
    nn = L.Add()([n, nn])
    n = nn


out = L.Conv2D(3, (1, 1), padding='same', name='out')(n)
```

```python
inp_low = L.Input([H, W, C])
inp_high = L.Input([high_H, high_W, high_C])

dense_low = L.Dense(32)(inp_low)
dense_high = L.Dense(32)(inp_high)
pool0 = L.MaxPool2D()(dense_high)
cat = L.Subtract()([inp_low, pool0])

c = L.Conv2D(64, (3, 3), activation='relu', padding='same')(cat)

b = L.MaxPool2D()(c)  #BatchNormalization(gamma_initializer=g_init)(c)
c = L.Conv2D(64, (3, 3), activation='relu', padding='same')(b)

c = L.Conv2D(64, (3, 3), activation='relu', padding='same')(c)

c = L.Conv2D(64, (3, 3), activation='relu', padding='same')(c)

c = L.Conv2D(64, (3, 3), activation='relu', padding='same')(c)

c = L.Conv2D(64, (3, 3), activation='relu', padding='same')(c)

c = L.Conv2D(64, (3, 3), activation='relu', padding='same')(c)

#b = L.BatchNormalization(gamma_initializer=g_init)(d)
d = L.Dense(32, activation='tanh')(c)

"""
conv1 = L.Conv2D(32, 3,activation='elu')(cat)
conv2 = L.Conv2D(32, 3,activation='elu')(conv1)
pool1 = L.MaxPool2D()(conv2)
conv3 = L.Conv2D(32, 3,activation='elu')(pool1)
conv4 = L.Conv2D(64,kernel_size=(5,5),activation='elu')(conv3)
"""
#flat = L.Flatten()(conv4)
#dense1 = L.Dense(256,activation='tanh')(flat)
#dense1 = L.Dense(32,activation='tanh')(conv4)
#dense2 = L.Dense(2,activation=tf.nn.log_softmax)(dense1)
#res = L.Reshape((32, 32, 2))(dense1)
out = L.Dense(2, activation=tf.nn.log_softmax, name="out")(d)
```

```python
discriminator = keras.models.Model([inp_low,inp_high], out, name="discriminator")
```

```python
discriminator.summary()
```

```python
noise = tf.placeholder('float32',[None,H,W,C])
real_data = tf.placeholder('float32',[None,high_H,high_W,high_C])


logp_real = discriminator([noise, real_data])

generated_data = generator(noise)

logp_gen = discriminator([noise, generated_data])
```

```python
d_loss = -tf.reduce_mean(logp_real[:,:,:,1] + logp_gen[:,:,:,0])

#regularize
d_loss += tf.reduce_mean(discriminator.layers[-1].kernel**2)

#optimize
disc_optimizer =  tf.train.GradientDescentOptimizer(1e-3).minimize(d_loss,var_list=discriminator.trainable_weights)
```

```python
g_loss = -tf.reduce_mean(logp_gen)

gen_optimizer = tf.train.AdamOptimizer(1e-4).minimize(g_loss,var_list=generator.trainable_weights)
```

```python
s.run(tf.global_variables_initializer())
```

```python
def sample_noise_batch(idxs):
    return X_low[idxs]

def sample_data_batch(idxs):
    return X_high[idxs]

#def sample_both_batch(bsize):
#    idxs = np.random.choice(np.arange(X_high.shape[0]), size=bsize)
#    return X_low[idxs], X_high[idxs]

def sample_images(nrow, ncol, sharp=False):
    size=np.random.choice(np.arange(X_low.shape[0]), size=nrow*ncol)
    images = generator.predict(sample_noise_batch(size))
    if np.var(images)!=0:
        images = images.clip(np.min(X_high),np.max(X_high))
    for i in range(nrow*ncol):
        plt.subplot(nrow,ncol,i+1)
        if sharp:
            plt.imshow(images[i].reshape((high_H, high_W, high_C)),cmap="gray", interpolation="none")
        else:
            plt.imshow(images[i].reshape((high_H, high_W, high_C)),cmap="gray")
    plt.show()

def sample_probas(bsize):
    bsize = np.random.choice(np.arange(X_low.shape[0]), size=bsize)
    plt.title('Generated vs real data')
    plt.hist(np.exp(np.mean(discriminator.predict([sample_noise_batch(bsize), sample_data_batch(bsize)])[:,:,:,1])),
             label='D(x)', alpha=0.5,range=[0,1])
    plt.hist(np.exp(np.mean(discriminator.predict([sample_noise_batch(bsize), generator.predict(sample_noise_batch(bsize))])[:,:,:,1])),
             label='D(G(z))',alpha=0.5,range=[0,1])
    plt.legend(loc='best')
    plt.show()
    
```

```python
from IPython import display
from tqdm import tnrange
```

```python
for epoch in tnrange(1):
    idxs = np.random.choice(np.arange(X_low.shape[0]), size=100)
    
    feed_dict = {
        real_data:sample_data_batch(idxs),
        noise:sample_noise_batch(idxs)
    }
    
    for i in range(20):
        s.run(disc_optimizer,feed_dict)
    
    for i in range(1):
        summary = s.run(gen_optimizer,feed_dict)
    
    display.clear_output(wait=True)
    plt.imshow(s.run(generator(tf.placeholder_with_default(X_high[:1],shape=X_high[:1].shape)))[0]) #.clip(0,1))
    plt.show()
    if epoch %10==0:
        display.clear_output(wait=True)
        sample_images(2,3,True)
        sample_probas(100)
        plt.show()
```
